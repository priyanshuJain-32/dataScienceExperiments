```python
import json # convert API response to json format

from urllib.parse import urlencode

import requests # to get the webpage
from bs4 import BeautifulSoup # to parse the webpage

import pandas as pd
import re # regular expression operators

from datetime import datetime
```

```python
response = requests.get("https://www.bbc.com/weather/1277333")
soup = BeautifulSoup(response.content,'html.parser')
```

```python
daily_high_values = soup.find_all('span', attrs={'class': 'wr-day-temperature__high-value'})
daily_low_values = soup.find_all('span', attrs={'class': 'wr-day-temperature__low-value'})
daily_summary = soup.find_all('div', attrs={'class': 'wr-day-summary'})
```

```python
daily_high_values_C = [daily_high_values[i].text.strip()[:2] for i in range(len(daily_high_values))]
daily_low_values_C = [daily_low_values[i].text.strip()[:2] for i in range(len(daily_low_values))]
daily_summary_list = re.findall('[A-Z][^A-Z]*',daily_summary[0].text.strip())
```

```python
dates = [str(datetime.today().year)+"/"+str(datetime.today().month)+"/"+str(datetime.today().day+i) for i in range(14)]
```

```python
data = pd.DataFrame({'Date':dates,'High (C)': daily_high_values_C, 'Low (C)' : daily_low_values_C,'Range (C)':[int(daily_high_values_C[i])-int(daily_low_values_C[i]) for i in range(len(daily_high_values_C))], 'Summary':daily_summary_list})
data
```
